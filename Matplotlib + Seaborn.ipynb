## Matplotlib example workflow

```python
#0. import matplotlib and get it ready for ploting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. setup plot
fig, ax=plt.subplots(figsize=(10,10)) #(width,height)

#3. plot data
ax.plot(x,y)

#4. customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5. save & show
fig.savefig("Graphs/sample_plot.png")
```

## Two options for subplots

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))
# plot to each different axis
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which onw should you use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
#len(over_50) 
over_50.head()
```

## pyplot method


```python
over_50.plot(figsize=(10,10), x='age',
            y='chol',
            c='target',
            kind='scatter');
```

## OO method mixed with pyplot method


```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#ax.set_xlim([45,100]);
```

## OO method from scratch

```python
fig, ax=plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"]);  

# customize the data
ax.set(title=" Heart Disease & Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol,age,thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);

# add data to ax0
ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
ax0.set(title=" Heart Disease & Cholestrol Levels",
            ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add data to ax1
ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
ax1.set(title=" Heart Disease & Max Heart Rate",
      xlabel="Age",
      ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# add title to the fig

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# set the style
plt.style.use('seaborn-whitegrid')

fig, ax=plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"],
                  cmap="winter");  #changes color scheme

# customize the data
ax.set(title=" Heart Disease & Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# customizing the y and x axis limitations
# subplot of chol,age,thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);

# add data to ax0
ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")
ax0.set(title=" Heart Disease & Cholestrol Levels",
            ylabel="Cholestrol")

# change the x axis lim
ax0.set_xlim([50,80])

ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add data to ax1
ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")
ax1.set(title=" Heart Disease & Max Heart Rate",
      xlabel="Age",
      ylabel="Max Heart Rate")

# change the y axis lim
ax1.set_ylim([60,200])

ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# add title to the fig

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
